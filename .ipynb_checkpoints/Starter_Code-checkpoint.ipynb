## Preprocessing

```python
from sklearn.preprocessing   import StandardScaler
from sklearn.model_selection import train_test_split
import pandas     as pd
import sklearn    as skl
import tensorflow as tf
import numpy      as np
```

```python
# Read the mouse data and the study results
charity_metadata_path = "charity_data.csv"
charity_meta   = pd.read_csv(charity_metadata_path)
charity_df     = pd.DataFrame(charity_meta)
charity_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

charity_df = charity_df.drop(axis=1, columns=['EIN','NAME'])
```

```python
# Determine the number of unique values in each column.
print(charity_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts = charity_df['APPLICATION_TYPE'].value_counts()
print(counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = list(counts[counts<500].index)

# Replace in dataframe
for app in application_types_to_replace:
    charity_df['APPLICATION_TYPE'] = charity_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
charity_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
bins_num = charity_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

bins_num = bins_num[bins_num>1]
bins_num
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list (bins_num[bins_num<100].index)
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    charity_df['CLASSIFICATION'] = charity_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
charity_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

charity_df = pd.get_dummies(charity_df,dtype=float)
charity_df.head()
```

```python
# Split our preprocessed data into our features and target arrays

y = charity_df['IS_SUCCESSFUL'].values
y

X = charity_df.drop('IS_SUCCESSFUL', axis=1).values
X

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split( X, y, random_state = 38)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])

layer_1 = 7
layer_2 = 14
layer_3 = 21

dnn = tf.keras.models.Sequential()

dnn.add(tf.keras.layers.Dense(units = layer_1, 
                              input_dim = number_input_features, 
                              activation='relu'))

dnn.add(tf.keras.layers.Dense(units = layer_2, 
                              activation = 'relu'))

#nn.add(tf.keras.layers.Dense(units = layer_3, 
#                             activation='relu'))

dnn.add(tf.keras.layers.Dense(units=1, 
                              activation='sigmoid'))

dnn.summary()
```

```python
# Compile the model
dnn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = dnn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = dnn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
